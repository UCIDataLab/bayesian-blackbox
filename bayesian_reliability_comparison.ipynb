```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import os
import csv
import pandas as pd



COLUMN_WIDTH = 3.25  # Inches
TEXT_WIDTH = 6.299213  # Inches
GOLDEN_RATIO = 1.61803398875
DPI = 300
FONT_SIZE = 8


def plot_reliability_diagram_error(N_list, filename_bayesian, filename_frequentist, figname):
    output = filename_bayesian.split('/')[-1].split('_')
    print(filename_bayesian)
    print(filename_frequentist)
    if output[2] == 'imagenet2':
        output = [output[0], 'imagenet2_topimages', output[4]]
    else:
        output = [output[i] for i in [0, 2, 3]]
    bayesian_estimation_error = np.genfromtxt(filename_bayesian, delimiter=',')[:,:3]
    frequentist_estimation_error = np.genfromtxt(filename_frequentist, delimiter=',')[:,:3]
    print(bayesian_estimation_error)
    print(frequentist_estimation_error)
    
    error_difference = (bayesian_estimation_error - frequentist_estimation_error) * 100. / frequentist_estimation_error
    plt.errorbar(N_list, error_difference.mean(axis=0), error_difference.std(axis=0),
                linestyle='None', marker='*', label='Frequentist');
    plt.savefig(figname, format='pdf', dpi=300, bbox_inches='tight');
    # return a list 
    for idx in range(3):
        output.append(round(error_difference[:,idx].mean(), 2))
    return output
```

```python
dir = 'output/accuracy_estimation_error'
filelist = [_ for _ in os.listdir(dir) if _[-3:]=='csv']
filelist.sort()
output_table = []
with open(os.path.join(dir, 'n_list'), 'r') as f:
    reader = csv.reader(f)
    N_list = [int(_) for _ in list(reader)[0]][:3]
    
for i in range(int(len(filelist)/2)):
    filename_bayesian = os.path.join(dir, filelist[2*i])
    filename_frequentist = os.path.join(dir, filelist[2*i+1])
    figname = 'figures/accuracy_estimation_error/' + '_'.join(filelist[2*i].split('_')[:-1]) + '.pdf'
    dataset = filelist[2*i].split("_")[2]
    output_table.append(plot_reliability_diagram_error(N_list, filename_bayesian, filename_frequentist, figname))
```

```python
output_filename = "output/calibration_error_comparison.csv"
pd.DataFrame(output_table).to_csv(output_filename, index=False, header=False)
```
